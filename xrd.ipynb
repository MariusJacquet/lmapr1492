## Analyse du spectre de diffraction infrarouge du cristal $Rb_2O$

```python
from pymatgen.ext.matproj import MPRester
from pymatgen.analysis.diffraction.xrd import XRDCalculator
```

```python
with MPRester("eydjkyvaohvx0D8I") as m:
    structure = m.get_structure_by_material_id("mp-1394")
```

```python
xrd = XRDCalculator("CuKa1")
pattern = xrd.get_pattern(structure)
```

J'ai choisi Cuka1 car c'est la valeur la plus proche de celle demandée dans l'énoncé. En effet, la valeur de longueur d'onde demandée est de $\lambda = 1.54060 \overset{\circ}{A}$ et CuKa1 vaut $1.541 \overset{\circ}{A}$

### Calcul des 3 premiers pics du diffractogramme

```python
i = 0
while i < 3:
    print("Pic du diffractogramme n°",i+1,"\n \n",
         "Angle 2Θ entre les 2 rayons :", pattern.x[i],"\n",
         "Intensité du premier pic :", pattern.y[i],"\n",
         "Indices hkl associés :", pattern.hkls[i],"\n",
         "Distance associée :", pattern.d_hkls[i],"\n",
         "\n-------------------------------------------------------------------------------------------------------------------------------- \n")
    i+=1
```

### Représentation graphique du diffractogramme du cristal limité aux 3 premiers pics

```python
xrd.show_plot(structure, two_theta_range=(0,37))
```
