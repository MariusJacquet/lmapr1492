## Symétrie des atomes de la molécules $Rb_2O$

```python
from pymatgen.ext.matproj import MPRester
from pymatgen.io.cif import CifWriter
from pymatgen.symmetry.analyzer import *
from pymatgen.io.cif import *
from pymatgen.core.operations import *
from pymatgen.symmetry.site_symmetries import *
import matplotlib.pyplot as plt
import matplotlib.image as img
```

```python
structure = MPRester("eydjkyvaohvx0D8I").get_structure_by_material_id("mp-1394")
CifWriter(structure,0.1, write_magmoms=False).write_file("mp-1394_symmetrized.cif")
```

```python
#Opérations de symétrie 
Sym1 = SymmOp.from_xyz_string('y, x, z')
Sym2 = SymmOp.from_xyz_string('z, x, y')
Sym3 = SymmOp.from_xyz_string('x+0.5, y+0.5, z')

#Coordonnées initiales des atomes
Rb1 = [0.75,0.25,0.25]
O1 = [0.5,0.0,0.5]
Rb2 = [0.25,0.25,0.75]

#Coordonées après symétrie des atomes
Rb1res = Sym1.operate(Rb1)
O1res = Sym2.operate(O1)
Rb2res = Sym3.operate(Rb2)
```

# Première symétrie

```python
print("La première symétrie agit sur l'atome de rubidium situé en", Rb1)
print("L'opération de symétrie éffectuée est [", Sym1.as_xyz_string(), "] et donne comme nouvelles coordonées", Rb1res, "pour l'atome de rubidium \nIl s'agit d'un plan miroir. \n")
print("Les matrices de transformations sont : \n", Sym1)

print("\n Ci-dessous, une représentation de la transformation effectuée.")
image_1=img.imread('Rb1.png')
fig, axes = plt.subplots(figsize=(10,10))
axes.get_xaxis().set_visible(False)
axes.get_yaxis().set_visible(False)
axes.imshow(image_1)
plt.show()
```

# Deuxième symétrie

```python
print("La deuxième symétrie agit sur l'atome d'oxygène' situé en", O1)
print("L'opération de symétrie éffectuée est [", Sym2.as_xyz_string(), "] et donne comme nouvelles coordonées", O1res, "pour l'atome d'oxygène' \nIl s'agit d'une rotation C3. \n")
print("Les matrices de transformations sont : \n", Sym2)

print("\n Ci-dessous, une représentation de la transformation effectuée.")

image_1=img.imread('O1.png')
fig, axes = plt.subplots(figsize=(10,10))
axes.get_xaxis().set_visible(False)
axes.get_yaxis().set_visible(False)
axes.imshow(image_1)
plt.show()
```

# Troisième symétrie

```python
print("La première symétrie agit sur l'atome de rubidium situé en", Rb2)
print("L'opération de symétrie éffectuée est [", Sym3.as_xyz_string(), "] et donne comme nouvelles coordonées", Rb2res, "pour l'atome de rubidium. \nIl s'agit d'une translation.\n")
print("Les matrices de transformations sont : \n", Sym3)

print("\n Ci-dessous, une représentation de la transformation effectuée.")

image_1=img.imread('Rb2.png')
fig, axes = plt.subplots(figsize=(10,10))
axes.get_xaxis().set_visible(False)
axes.get_yaxis().set_visible(False)
axes.imshow(image_1)
plt.show()
```
