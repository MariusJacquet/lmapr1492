```python
from pymatgen.ext.matproj import MPRester
from pymatgen.io.cif import CifWriter
from pymatgen.symmetry.analyzer import *
from pymatgen.io.cif import *
from pymatgen.core.operations import *
from tabulate import tabulate
from pymatgen.symmetry.site_symmetries import *
import matplotlib.pyplot as plt
import matplotlib.image as img
```

```python
structure = MPRester("eydjkyvaohvx0D8I").get_structure_by_material_id("mp-30460")
CifWriter(structure,0.1, write_magmoms=False).write_file("mp-30460_symmetrized.cif")
```

```python
a = SpacegroupAnalyzer(structure).get_symmetry_operations()
Sym1 = SymmOp.from_xyz_string('y, -x, -z')
Sym2 = SymmOp.from_xyz_string('z, x, y')
Sym3 = SymmOp.from_xyz_string('z, y, x')
```

```python
Bi = [0.0,0.5,0.5]
Y = [0.0,0.0,0.5]
Ni = [0.75,0.25,0.75]
```

```python
Bires = Sym1.operate(Bi)
Yres = Sym2.operate(Y)
Nires = Sym3.operate(Si)
```

```python
print(tabulate([
	["("+str(Sym1.as_xyz_string()+")"),str(Bi),str(Bires)],
	["----------------------------------","",""],
	[Sym1,"",""],
	["----------------------------------","-----------------","----------------------------------"],
	["("+str(Sym2.as_xyz_string()+")"),str(Y),str(Yres)],
	["----------------------------------","",""],	
	[Sym2,"",""],
	["----------------------------------","-----------------","----------------------------------"],
	["("+str(Sym3.as_xyz_string()+")"),str(Ni),str(Nires)],
	["----------------------------------","",""],	
	[Sym3,"",""],
	["----------------------------------","-----------------","----------------------------------"]],
	headers=["Opération de symétrie","Atome de départ","Atome d'arrivée"]))
print("\n")
```
